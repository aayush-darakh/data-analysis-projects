```python
# create a root variable
root = "https://www.football-data.co.uk/mmz4281/"
```

## 2.1 Multiple leagues


* https://www.football-data.co.uk/mmz4281/2526/E0.csv
* https://www.football-data.co.uk/mmz4281/2526/E1.csv
* https://www.football-data.co.uk/mmz4281/2526/E2.csv
* https://www.football-data.co.uk/mmz4281/2526/E3.csv
* https://www.football-data.co.uk/mmz4281/2526/EC.csv




```python
# creating a list of leagues
leagues = ['E0', 'E1', 'E2','E3','EC']
frames = []

#looping through leagues, read, multiple csv and append it into the list
for league in leagues:
  df = pd.read_csv(root + '2526' + '/' + league + '.csv')
  frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st, 2nd and 3rd Element
frames[0]
```

## 2.2 Multiple Seasons

```python
for seasons in range(15,25,1):
  print(str(seasons)+str(seasons+1))
```

```python
# creating a list of leagues and an empty list to store all dataframes
leagues = ['E0', 'E1', 'E2']
frames_s = []

# looping through all leagues from 1st loop and range of season through 2nd loop
for league in leagues:
  for season in range(15,25,1):
    df_s = pd.read_csv(root + str(season)+str(season+1) + "/" + league + ".csv")
    df_s.insert(1,'Season',season)
    frames_s.append(df_s)
```

```python
# total frames generated 3*10 = 30
len(frames_s)
```

# 3)Organize all the data in the Dictionary

```python
# create a dictionary with original name of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1',
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
# get a dictionary element
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
  print(dict_countries[league])
```

```python
#creating new dictionary whose keys stores all leagues and values have data from year 2015-2024(2nd for loop) all concatened in one df respectively
dict_data_history = {}

for league in dict_countries:
  frames_D = []
  for seasons in range(15,25,1):
    df_D = pd.read_csv(root + str(seasons) + str(seasons+1) + "/" + dict_countries[league] + ".csv")
    df_D.insert(1,'Season',seasons)
    frames_D.append(df_D)
  df_concat = pd.concat(frames_D)
  dict_data_history[league] = df_concat
```
