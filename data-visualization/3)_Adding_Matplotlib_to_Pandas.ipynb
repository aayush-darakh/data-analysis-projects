```python
# importing matplotlib and pandas
import pandas as pd
import matplotlib.pylab as plt
```

```python
# reading csv file
df_pr = pd.read_csv("population_total.csv")
df_pr.dropna(inplace = True)
df_pivot = df_pr.pivot(index = 'year', columns = 'country', values = 'population' )
df_pivot = df_pivot[['China', 'United States', 'India', 'Germany', 'Brazil']]
df_pivot
```

# 2.Lineplot

```python
df_pivot.plot(kind='line',
              xlabel = 'Year',
              ylabel = 'Population',
              title = 'Population  (1955- 2020)',
              figsize = (8,4))

#save plot
plt.savefig('my_test.png')
#show plot
plt.show()
```
