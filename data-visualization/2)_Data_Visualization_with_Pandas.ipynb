```python
import pandas as pd
```

```python
# read csv file
df_pr = pd.read_csv("population_total.csv")
```

```python
# first look at dataset
df_pr
```

# 1.Making pivot table
.pivot(): returns reshaped dataframe organized by give index/column values ("Pivot without aggregation)

```python
# dropping null values using .dropna() method
df_pr.dropna(inplace = True)
df_pr
```

```python
# making a pivot table
df_pivot = df_pr.pivot(index = "year", columns = "country", values = "population")
df_pivot
```

```python
# select some countries
df_pivot = df_pivot[['China', 'United States', 'India', 'Germany', 'Brazil']]
```

```python
# showing pivot table
df_pivot
```

# 2.Lineplot

```python
df_pivot.plot(kind='line', xlabel = 'Year', ylabel = 'Population', title = 'Population  (1955- 2020)', figsize = (8,4))
```

# 3.Barplot
(note : put text data in the index)

## 3.1 Single Barplot

```python
# selecting only one year (2015)
df_pivot_2015 = df_pivot[df_pivot.index.isin([2015])]
```

```python
# showing new Dataframe
df_pivot_2015
```

```python
# transpose dataframe from rows to columns or viceversa (.T)
df_pivot_2015 = df_pivot_2015.T
df_pivot_2015.T
```

```python
# making barplot
df_pivot_2015.plot(kind = 'bar', color = 'pink',xlabel = 'Year', ylabel = 'Population', title = 'Population 2015', figsize = (8,4))
```

## 3.2 Barplot grouped by 'n' variables

```python
# selecting few years
df_pivot_sample = df_pivot[df_pivot.index.isin([1975,1985,1995,2005,2015])]
```

```python
# making a group barplot
df_pivot_sample.plot(kind = 'bar',xlabel = 'Year', ylabel = 'Population', figsize = (8,4))
```

# 4.Piechart
using same dataframe as barplot as it follows the same logic

```python
# changing column name as it's not good for practice to have numbers(int etc) in columns
# .rename() method is used to rename column
df_pivot_2015.rename(columns={2015.0:'2015'}, inplace = True)
```

```python
# making a plot
df_pivot_2015.plot(kind = 'pie', y='2015', title = 'Population in 2015(%)')
```
