```python
import pandas as pd
```

```python
# reading a csv file
df_exams = pd.read_csv('student_performance.csv')
df_exams.head()
```

# 1 Create Index

```python
import numpy as np
import random as rd
```

```python
# creating non repetitive values for index
new_index = np.arange(0,500,1)
```

```python
# shuffling the indexes
rd.shuffle(new_index)
```

```python
# create a new column with new index
df_exams['New_Index'] = new_index
df_exams
```

# 2 Set Index

```python
# set New_Index column as index
df_exams.set_index('New_Index', inplace = True)
df_exams
```

# 3 Sort Index


```python
# sort dataframe by New_Index
df_exams.sort_index()
```

```python
# sort descending
df_exams.sort_index(ascending=False, inplace=True)
df_exams
```
