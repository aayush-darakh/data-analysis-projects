### 1.3 using assign() method

```python
# create random numbers for new score column
score1 = np.random.randint(1,100, size = 500)
score2 = np.random.randint(1,100, size = 500)
```

```python
# create series using random integers
series1 = pd.Series(score1, index=np.arange(0,500,1))
series2 = pd.Series(score2, index=np.arange(0,500,1))
```

```python
# using assign() method to add multiple columns
df_exams = df_exams.assign(Score_1 = series1, Score_2 = series2)
# assign() creates copy of the dataframe, to update we need to overwrite the dataframe which is done in the above statemtent
```

```python
# show dataframe
df_exams
```

### using insert() method


```python
# using index() method to add column in specific location
# index() doesn't create copy, it updates the values (not like assign())
df_exams.insert(3, 'Test', series1)
```

```python
#show dataframe
df_exams
```
