```python
import pandas as pd
import numpy as np
```

```python
# reading csv file
df_exams = pd.read_csv("student_performance.csv")
df_exams.head()
```

# 1 Math Operations

## 1.1 Operations in Columns

```python
# select a column and calculate sum
df_exams['Midterm_Score'].sum().item()
```

```python
# calculate count, mean, std, max and min
print(df_exams['Midterm_Score'].mean())
print(df_exams['Midterm_Score'].count())
print(df_exams['Midterm_Score'].std())
print(df_exams['Midterm_Score'].max())
df_exams['Midterm_Score'].min()
```

```python
# using .describe()
df_exams.describe()
```

## 1.2 Operations in Rows

```python
# calculating the sum in a row
df_exams['Assignments_Completed'] +	df_exams['Midterm_Score'] +	df_exams['Final_Score']
```

```python
# calculating avg score and assigining it to a new column
df_exams["average"] = (df_exams['Assignments_Completed'] +	df_exams['Midterm_Score'] +	df_exams['Final_Score'])/3
```

```python
# showing dataframe, .round(n) round-off upto n digits
df_exams.round(2)
```
