```python
import pandas as pd
import numpy as np
```

## 1 Creating a Dataframe from an Arrays

### option 1

```python
# creating an Array
data = np.array([[1,2],[3,4],[5,6]])
```

```python
#creating an dataframe
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'], #index represents row name
                  columns = ['column1', 'columns2']) #column represent column name
```

```python
#showing dataframe
df
```

### option 2 (list)

```python
#creating a list
data = [[1,2],[3,4],[5,6]]
```

```python
#creating an dataframe
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'], #index represents row name
                  columns = ['column1', 'columns2']) #column represent column name
```

```python
#showing dataframe
df
```

## 2 Creating a Dataframe from Dictionary

```python
#creating a list
states = ['Maharashtra', 'Rajasthan', 'Gujarat']
population = [20358465, 57486543, 6847365]
```

```python
#storing lists within dictionary
dict_s = {'States': states, 'Population': population}
```

```python
#creating a dataframe
df_p = pd.DataFrame(dict_s)
```

```python
#showing dataframe
df_p
```
