```python
import pandas as pd
```

```python
# reading csv file
df_exams = pd.read_csv('student_performance.csv')
df_exams.head()
```

# 1 Value Counts method

```python
# counting gender elements

# len function
len(df_exams['Gender'])

# .count() method
df_exams['Gender'].count().item()
```

```python
# counting elements by category
df_exams['Gender'].value_counts()
```

```python
# return the relative frequency (divide all values by the sum of all values)
df_exams['Gender'].value_counts(normalize=True) #gives pecentage of each category
```

```python
# counting "Parental_Level_of_Education" elements by category
df_exams['Parental_Level_of_Education'].value_counts()
```

```python
# returns the the relative frequency and round to 2 decimal places
df_exams['Parental_Level_of_Education'].value_counts(normalize=True).round(2)
```
