```python
import pandas as pd
```

```python
# reading a csv file
df_exams = pd.read_csv('student_performance.csv')
df_exams.head()
```

# 1 Rename Column

```python
# rename column and overwrite values of the dataframe
df_exams = df_exams.rename(columns = {'Gender': 'gender'})
df_exams
```

```python
# rename multiple columns and update the dataframe with inplace argument
df_exams.rename(columns = {
    'Age' : 'age',
    'Study_Hours_per_Week' : 'study_hours_per_week',
    'Attendance_Percentage' : 'attendance_percentage'
}, inplace = True)
```

```python
# shoe dataframe
df_exams
```

# 2 Rename using Index argument

```python
# rename index 0,1 and 2 and update the df
df_exams.rename(index={0:'A', 1:'B', 2:'C', 3:'D'}, inplace = True)
```
