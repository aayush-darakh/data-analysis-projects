```python
import pandas as pd
```

```python
# reading a csv file
df_exams = pd.read_csv('student_performance.csv')
df_exams.head()
```

# 1 Sort a dataframe

```python
# sort by one column
df_exams.sort_values(by='Midterm_Score') # can exclude by keyword
```

```python
# sort descending by one column
df_exams.sort_values('Midterm_Score', ascending=False)
```

```python
# sort descending by multiple columns
df_exams.sort_values(['Final_Score', 'Midterm_Score'], ascending=False)
```

```python
# sort descending by multiple columns and update dataframe
df_exams.sort_values(['Final_Score', 'Midterm_Score'], ascending=False, inplace=True)
df_exams #checks final score first, if matches then considers mideterm score
```

```python
# sort descending with a key function
df_exams.sort_values('Parental_Level_of_Education', ascending = True, key = lambda col: col.str.lower())
```
