```python
import pandas as pd
```

## 1.5 Exclusive Left Join
to create anytype of exclusive join, always start with how = 'outer'

```python
# create outer join
df1.merge(df2, on = 'id', how = 'outer')
```

```python
# exclusive outer join, indicator = True
df1.merge(df2, on = 'id', how = 'outer', indicator = True)
```

```python
# query to create exclusvie left join
df_exclusive_left = df1.merge(df2, on = 'id', how = 'outer', indicator = True).query("_merge == 'left_only'")
df_exclusive_left
```

```python
# shape
df_exclusive_left.shape
```

### 1.5.1 Exercise

```python
# make a copy of the df_movies dataframe
df_movie_2 = df_movies.copy()
df_movie_2
```

```python
# set the first 1000 values of 'imdb_title_id' column as 'tt1234567890'
for index in df_movie_2.index:
  if index < 1000:
    df_movie_2.iloc[index, 0] = 'tt1234567890'
```

```python
df_movie_2
```

```python
df_ratings.columns
```

```python
# merge df_movie_2 and df_ratings (exclusive left join)
df_movie_2.merge(df_ratings, on = 'imdb_title_id', how = 'outer', indicator = True)
```

```python
df_exclusive_left = df_movie_2.merge(df_ratings, on = 'imdb_title_id', how = 'outer', indicator = True).query("_merge == 'left_only'")
df_exclusive_left
```

```python
# shape
print(df_exclusive_left.shape)
print(df_movie_2.shape)
print(df_ratings.shape)

# got 1000 because we created exclusve id above for first 1000 rows which we updated above
```

## 1.7 Exclusive Right Join

```python
# query to crate exclusive right join
df1.merge(df2, on='id', how='outer', indicator=True).query("_merge == 'right_only'")
```

### 1.7.1 Exercise

```python
# make a copy of the df_ratings dataframe
df_ratings_2 = df_ratings.copy()
```

```python
# set the first thousand values of 'imdb_title_id' columns as 'tt1234567890'
for index in df_ratings_2.index:
  if index < 1000:
    df_ratings_2.loc[index, 'imdb_title_id'] = 'tt1234567890'
```

```python
# merge df_movies and df_ratings_2 (exclusive right)
df_movies.merge(df_ratings_2, on='imdb_title_id', how='outer', indicator = True)
df_exclusive_right = df_movies.merge(df_ratings_2, on='imdb_title_id', how='outer', indicator = True).query("_merge == 'right_only'")
df_exclusive_right.value_counts('_merge')
```

```python
# shape
print(df_ratings_2.shape)
print(df_exclusive_right.shape)
print(df_movies.shape)
```
