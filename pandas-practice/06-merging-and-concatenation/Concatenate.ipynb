```python
import pandas as pd
```

```python
df_movies = pd.read_csv('IMDb movies.csv', engine='python', on_bad_lines = 'skip')
df_ratings = pd.read_csv('IMDb ratings.csv')
```

```python
df_movies = df_movies[['imdb_title_id', 'title', 'year','genre', 'country']]
df_ratings = df_ratings[['imdb_title_id', 'total_votes', 'mean_vote']]
```

# 1 concat()

## 1.1 Concatenate Vertically

```python
df1 = pd.DataFrame({
    'id' : ['A', 'B', 'C', 'D'],
    'age' : [30, 23, 32, 22]
})

df2 = pd.DataFrame({
    'id' : ['E', 'F', 'G', 'H'],
    'age' : [40, 21, 19, 24]
})
```

```python
pd.concat([df1, df2], axis = 0, ignore_index = True)
```

### 1.1.1 exercise

```python
# exctract a 50% sample of the original dataframe (df_movies)
df_sample = df_movies.sample(frac = 0.5)
```

```python
# shape of dataframes that we'll concatenate
print(df_movies.shape)
df_sample.shape
```

```python
# concatenate df_movies and df_sample (vertically along the rows)
df_concat_vertically = pd.concat([df_movies, df_sample], axis = 0)
```

```python
# shape
df_concat_vertically.shape
```

## 1.2 Concatenate Horizontally

```python
df1 = pd.DataFrame({
    'id' : ['A', 'B', 'C', 'D'],
    'age' : [30, 23, 25, 22]
})

df2 = pd.DataFrame({
    'job' : ['Doctor', 'Statistician', 'Accountant', 'Developer'],
})
```

```python
pd.concat([df1, df2], axis = 1)
```

### 1.2.1 exercise

```python
# shape of dataframes that we'll concatenate
print(df_movies.shape)
df_ratings.shape
```

```python
df_movies.set_index('imdb_title_id', inplace = True)
df_ratings.set_index('imdb_title_id', inplace = True)
```

```python
# concatenate df_movies and df_ratings on 'imdb_title_id' (horizontally)

df_concat_horizontally = pd.concat([df_movies, df_ratings], axis = 1)
df_concat_horizontally
```

```python
# shape
df_concat_horizontally.shape
```
