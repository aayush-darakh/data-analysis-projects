```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace = True)
# select column
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe
df
```

# 1.Selecting with a Single Value
iloc[row_position, column_position]


```python
# get the height of   L.Messi
df.iloc[0, 3].item()
```

```python
# get the weight of Cristiano Ronaldo
df.iloc[1,4].item()
```

```python
# get all rows inside the 'height_cm' column
df.iloc[:,3]
```

```python
# get all columns that corrrespond to the index "Cristiano Ronaldo"
df.iloc[1, :]
```

# 2.Selecting with a list of values

```python
# get all data about 'L. Messi' and 'Cristiano Ronaldo'
df.iloc[[0,1], :]
```

```python
# get the height of 'L. Messi' and 'Cristiano Ronaldo'
df.iloc[[0,1], 3]
```

```python
# get the height and weight 'L. Messi'
df.iloc[0, [3,4]]
```

```python
# get the height and weight 'L. Messi' and 'Cristiano Ronaldo'
df.iloc[[0,1], [3,4]]
```

# 3.Select a range of data with a slice
Note that start is included but stop is excluded

```python
# slice columns labels age/club
# players = ['L. Messi', 'Cristiano Ronaldo']
players = [0,1]
df.iloc[players, 1:7]
```

```python
# slice index labels (top1 to top10 player)
# columns = ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1,2,3,4]

# get top1 and top10 player name
df.iloc[0:10, columns]
```

# 4.Selecting with Conditions
iloc cannot accept a boolean series, but only a boolean list. We have to use the list funciton to convert a series into a boolean list

```python
# one condition: select player with height above 180 cm
# columns = ['age', 'dob', 'height_cm', 'weight_kg']
columns = [1,2,3,4]

df.iloc[list(df['height_cm']>180), columns]
```

```python
# multiple conditions : select player with height above 180 cm and from Argentina
df.loc[((df['height_cm']>180) & (df['nationality'] == 'Argentina')), :]
```
