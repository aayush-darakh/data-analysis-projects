```python
import pandas as pd
```

```python
# read csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace = True)
# select columns
columns = ['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']
```

```python
# show df
df[columns]
```

# 1.sample()

```python
# extract 10 random elements from the "nationality" column
df['nationality'].sample(10, random_state = 69) #random_state generates the same numbers everytime
```

```python
# extract a random 20% sample of the dataframe
df.sample(frac=0.2,random_state = 25)
```

```python
# upsample: increase the sampling rate -> frac>1 (Note : replace parameter has to be True for frac parameter > 1)
df.sample(frac=2, replace = True, random_state = 69)
```
