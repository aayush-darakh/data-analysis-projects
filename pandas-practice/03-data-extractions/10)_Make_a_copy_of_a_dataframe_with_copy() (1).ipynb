```python
import pandas as pd
import numpy as np
```

```python
# read csv file
df = pd.read_csv('players_20.csv')
# set index
df.set_index('short_name', inplace = True)
# select columns
df = df[['long_name', 'age', 'dob', 'height_cm', 'weight_kg', 'nationality', 'club']]
```

```python
# show dataframe
df
```

# 1.copy()

## 1.1 deep = True
### copy doesnt get updated

```python
# deep = True by default (Modifications to the data or indices of the copy)
df_copy = df.copy()
```

```python
# update value in og dataframe
df.loc['Cristiano Ronaldo', 'height_cm'] = 190
```

```python
# copy vs og dataframe
df_copy # doesn't gets updated as it is independent of og df
df
```

## 1.2 deep = False
### (both gets updated)

```python
# Deep = False (Any changes to data of the og will be reflected in the shallow copy and vice versa)
df_shallow_copy = df.copy(deep=False)
```

```python
# update value in og df
df.loc['L. Messi', 'height_cm'] = 175
```

```python
# shallow copy vs og df
df
df_shallow_copy # both gets updated as they are dependent
```

### 1.3 Simple Assignment
#### (same as Deep = False)

```python
# make a copy
df_new_copy = df
```

```python
# update value in og df
df.loc['Neymar Jr', 'height_cm'] = 182
```

```python
# copy vs og df
df
df_new_copy # both df gets updated as it is a shallow copy
# same as using copy method where deep = False
```
