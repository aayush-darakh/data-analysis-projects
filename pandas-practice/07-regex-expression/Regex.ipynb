# The re module

```python
import re
```

## 1.1 re.search()

```python
text = """Hello World.
I love Python!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""
```

```python
# return the first match
re.search(r'\d', text)

# as 9 is the first didgit which occurred first
```

```python
# flags
re.search(r'^hello', text, flags = re.I) # if needed check all flags from the website
```

## 1.2 re.findall()

```python
# return all matches
re.findall(r"\d", text)
```

# Exercises

```python
      # Match only punctuations
text = """Hello World.
I love Python!!!!
My first lucky number is 987-654-321
My second lucky number is 876-543-210
My third lucky number is 765-432-100
My fourth lucky number is 123-456-123-123-456
"""

# [^ ] -> match any character that is NOT in brackets
#\s white space
#\w word character

re.findall(r'[^\w\s]', text)
```

```python
# Match right date format (MM-DD-YYYY)
text = '''
13-04-2021
2021-13-04
2021-04-13
'''
#\d -> represents digits
# {} -> exact number (to give shape)

re.findall(r'\d{2}-\d{2}-\d{4}', text)
```

```python
# Match right username format (4-14 characters and only letters or digits)
text = '''
username10
abc
10
'''

# [] -> matches character in the bracket with OR condition
# {min,max}

re.findall(r'[a-zA-Z0-9]{4,14}', text)
```
