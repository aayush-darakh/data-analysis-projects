```python
import pandas as pd
```

```python
# reading csv file
df_boston = pd.read_csv('Boston House Prices.csv')
df_boston
```

# 1 Multi Linear Regression

```python
import statsmodels.api as sm
```

## 1.1 Define independent and depenedent variable

```python
# dependent
y = df_boston['Value']
#independent
X = df_boston[['Rooms', 'Distance']]
```

## 1.2 Making the regression: Adding a constant and fitting the model

```python
# add a constant
X = sm.add_constant(X)

# fitting the model
lm = sm.OLS(y, X).fit()
```

```python
lm.summary()
```
