```python
import pandas as pd
import numpy as np
```

```python
# reading a csv file
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
# showe dataframe
df_laptops.head(5)
```

# 1.Create a Conditional Column for more than 2 choices: np.select()

```python
# create an array based on multiple price tiers (+2 choices)

# create conditions and value variables
conditions = [
    df_laptops['Price_euros'] > 3000,
    (df_laptops['Price_euros'] > 2000) & (df_laptops['Price_euros'] <= 3000),
    (df_laptops['Price_euros'] > 800) & (df_laptops['Price_euros'] <= 2000),
    df_laptops['Price_euros'] <= 800
]
values = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
# set it to new column
df_laptops['Price_tiers'] = np.select(conditions, values, default='Other')
```

```python
# show dataframe
df_laptops
```

```python
# counting elements under Price_tiers
df_laptops['Price_tiers'].value_counts()
```

```python
(df_laptops["Price_euros"] > 3000).value_counts()
```

```python
# create an array based on multiple screen size tiers (+2choices)
# Too big>16, big>14, small>12, too small<=12
# create condition and value variables
conditions = [
    df_laptops['Inches'] > 16,
    (df_laptops['Inches'] <= 16) & (df_laptops['Inches'] > 14),
    (df_laptops['Inches'] <= 14) & (df_laptops['Inches'] > 12),
    (df_laptops['Inches'] <= 12)
]

values = ['Too Big', 'Big', 'Small', "Too Small"]
```

```python
# set it to a new column
df_laptops['Screen_size'] = np.select(conditions, values, default='Other')
```

```python
# show dataframe
df_laptops
```

```python
# count values inside "screen_size" columns
df_laptops.value_counts('Screen_size')
```
