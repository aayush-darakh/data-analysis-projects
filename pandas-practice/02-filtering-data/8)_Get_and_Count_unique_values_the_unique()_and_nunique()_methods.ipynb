```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
df_laptops
```

# **1.unique()**
return unique values of Series objects. (Unique are returned in order of appearance.)
returns NumPy Array


```python
# get unique elements in "Company" column
df_laptops['Company'].unique()
```

```python
# get unique elements in "Inches" column
df_laptops['Inches'].unique()
```

```python
# get the lenght of unique elements
len(df_laptops['Company'].unique())
```

# **2.nunique() method**
Returns number of unique elements in the object(Excludes NA values by default)|



```python
# get unique elements in "Company" column
df_laptops['Company'].nunique()
```

```python
# get unique elements in "Inches" column
df_laptops['Inches'].nunique()
```
