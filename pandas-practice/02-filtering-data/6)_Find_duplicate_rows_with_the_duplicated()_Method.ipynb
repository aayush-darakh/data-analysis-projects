## 1.2 Example 2
Task : Find the cheapest and most expensive laptop each company using the sort_values and duplicated() methods (keep "first", "last" and False)

```python
# sort dataframe ascending by company and price
# cheapest first and most expensive last
df_laptops.sort_values(['Company', 'Price_euros'], inplace = True)
df_laptops
```

```python
# check all value categories in "Company" column
df_laptops.value_counts('Company')
```

```python
# duplicated values in "Company" columns (default -> keep='first)
duplicated_first = df_laptops['Company'].duplicated(keep = 'first')
duplicated_first
```

```python
# show dataframe with duplicated values in "Company" column
df_laptops[duplicated_first]

# keep='first' (cheapest laptop per company)
# show dataframe with one duplicated values in "Company" column
df_laptops[~duplicated_first][['Company', 'Price_euros']] # this list shows the only cheapest laptop of the particular company
```

```python
# keep = last
duplicated_last = df_laptops['Company'].duplicated(keep = 'last')
duplicated_last
```

```python
# keep = 'last' (most expensive laptop per company)
# show dataframe woth non duplicated values in "Company" column
df_laptops[~duplicated_last][['Company', 'Price_euros']]
```

```python
# keep = FALSE
duplicated_false = df_laptops['Company'].duplicated(keep = False)
duplicated_false
```

```python
# show dataframe woth non duplicated values in "Company" column
df_laptops[~duplicated_false]
```

```python
# all elements in company are repeated atleast once
df_laptops.value_counts("Company")
```
