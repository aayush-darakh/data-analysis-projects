```python
import pandas as pd
```

```python
df_laptops = pd.read_csv('laptop_price.csv')
df_laptops.head()
```

# 1.drop_duplicates()

```python
# drop duplicates in 2 or more columns
df_laptops.drop_duplicates('Company')[['Company', 'Price_euros']]
```

```python
# sort dataframe ascending by company and values
df_laptops.sort_values(['Company', 'Price_euros'], inplace=True)
# cheapest at first and most expensive at bottom
```

```python
# cheapest: keep = 'first' for dropping duplicate columns
df_laptops.drop_duplicates(['Company'], keep = 'first')[['Company', 'Price_euros']]
```

```python
# expensive: keep = 'last' for dropping duplicate columns
df_laptops.drop_duplicates(['Company'], keep = 'last', inplace = True, ignore_index = True)
```

```python
# Arguments:
# ignore_index = True : the resulting index will be replaced by 0,1,2..

df_laptops[['Company', 'Price_euros']]
```

# 2.Exercise 1
Task: Find the biggest and smallest screen size in the laptops of each company using the .sort_values() and .duplicated() methods (keep: "first", "last", False)

```python
df_l = pd.read_csv('laptop_price.csv')
df_l
```

```python
# sort dataframe ascending by company and inches
# smallest first and biggest last
df_l.sort_values(['Company', 'Inches'], inplace = True)
df_l
```

```python
# smallest : keep = 'first'
dupl_f = df_l.duplicated(['Company'], keep = 'first')
df_l[~dupl_f][['Company', 'Inches']]
```

```python
# smallest : keep = 'last'
dupl_l = df_l.duplicated(['Company'], keep = 'last')
df_l[~dupl_l][['Company', 'Inches']]
```
