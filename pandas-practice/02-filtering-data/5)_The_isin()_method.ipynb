```python
import pandas as pd
import numpy as np
```

```python
# reading a csv file
df_laptops = pd.read_csv('laptop_price.csv')
```

```python
# show dataframe
df_laptops.head()
```

# 1.isin() : Single Filtering
# is in method

```python
# select Apple or HP laptops
df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
 # filter dataframes and use value_counts
df_laptops[df_laptops['Company'].isin(['Apple', 'HP'])].value_counts('Company')
```

# 2.isin(): Multiple Filtering

```python
# find notebooks or ultrabooks from Apple or HP laptops
f1 = df_laptops['TypeName'].isin(['Notebook','Ultrabook'])
f2 = df_laptops['Company'].isin(['Apple', 'HP'])
```

```python
#filter dataframe
df_laptops[f1 & f2]
```

```python
# count
df_laptops[f1 & f2].value_counts('Company')
```
