# 1.Filter Dataframe based on one condition

```python
# find which rows have "Apple" in the "Company" column
df_laptops['Company'] == 'Apple'
```

```python
# filter dataframe based on condition
df_laptops[df_laptops["Company"] == "Apple"]
```

```python
# verify which companiea are in the "Company" with value counts
df_laptops[df_laptops["Company"] == "Apple"].value_counts("Company")
```
