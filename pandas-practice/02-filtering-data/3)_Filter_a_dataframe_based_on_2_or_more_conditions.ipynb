```python
import pandas as pd
import numpy as np
```

```python
# reading csv file
df_laptop = pd.read_csv('laptop_price.csv')
```

```python
# show df
df_laptop.head()
```

# 1.Filter dataframe based on 1 condition

```python
# find Apple laptops
df_laptop["Company"] == "Apple"
```

```python
# create an array based on price > 2000 euros
df_laptop["Price_euros"] > 2000
```

# 2.Filter dataframe based on multiple conditions

```python
# find apple laptops that cost more than 2000 euros
(df_laptop["Company"] == "Apple") & (df_laptop["Price_euros"] > 2000)
```

## 2.1 &(and)

```python
# filter dataframe based on multiple condition
df_laptop[(df_laptop["Company"] == "Apple") & (df_laptop["Price_euros"] > 2000)]
```

# 2.2 |(or)

```python
# find apple or dell laptops
(df_laptop['Company'] == 'Apple') | (df_laptop['Company'] == 'Dell')
```

```python
# filter dataframe based on multiple conditions
df_laptop[(df_laptop['Company'] == 'Apple') | (df_laptop['Company'] == 'Dell')].value_counts('Company')
```

```python
# find laptops from Apple or Dell that cost over 2000 Euros
((df_laptop['Company'] == 'Apple') | (df_laptop['Company'] == 'Dell')) & (df_laptop["Price_euros"]>2000)
```

```python
# filter dataframe based on multiple conditions
df_laptop[((df_laptop['Company'] == 'Apple') | (df_laptop['Company'] == 'Dell')) & (df_laptop["Price_euros"]>2000)]
```
