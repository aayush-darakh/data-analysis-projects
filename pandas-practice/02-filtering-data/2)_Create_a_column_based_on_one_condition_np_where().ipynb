# 1.Create a column based on 1 conditions: np.where()

```python
import numpy as np
```

```python
# create an array based on price tiers
np.where(df_laptop["Price_euros"] > 2000, 'Expensive', 'Cheap')
```

```python
# set it to new column
df_laptop['price_tier'] = np.where(df_laptop["Price_euros"] > 2000, 'Expensive', 'Cheap')
```

```python
#show first 5 rows
df_laptop.head(5)
```

```python
# count values in price_tier column
df_laptop.value_counts("price_tier")
```
